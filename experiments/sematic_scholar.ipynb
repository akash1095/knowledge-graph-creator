```python
class Paper:
    def __init__(
        self,
        paper_id,
        title,
        year,
        abstract,
        publicationtypes,
        publication_date,
        pdf,
        keywords,
    ):
        self.paper_id = paper_id
        self.title = title
        self.year = year
        self.abstract = abstract
        self.publication_types = publicationtypes
        self.publication_date = publication_date
        self.pdf = pdf
        self.keywords = keywords  # how to get keywords?


class Author:
    def __init__(self, author_id, name, affiliations):
        self.author_id = author_id
        self.name = name
        self.affiliations = affiliations


class Venue:
    def __init__(self, venue_id, name, venue_type, issn):
        self.venue_id = venue_id
        self.name = name
        self.venue_type = venue_type
        self.issn = issn
```
