```python
import os
from groq import Groq

# Initialize Groq client (reads API key from environment)
client = Groq(api_key=os.environ.get("GROP_API_KEY"))

# Make a simple request
chat_completion = client.chat.completions.create(
    messages=[{"role": "user", "content": "What is a Knowledge Graph?"}],
    model="llama-3.3-70b-versatile",  # Fast, powerful open-source model
)

# Print the response
print(chat_completion.choices[0].message.content)
```

```python
import requests
import os

api_key = os.environ.get("GROP_API_KEY")
url = "https://api.groq.com/openai/v1/models"

headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
for model_info in response.json()["data"]:
    print(
        f"Model ID: {model_info['id']}, Owned by: {model_info['owned_by']}, Context Length: {model_info['context_window']}"
    )
```
